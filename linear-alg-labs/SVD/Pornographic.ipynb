```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import numpy as np
```

```python
points = []

def f_1(x):
    return -1/20 * (x - 18) ** 2 + 20
def f_2(x):
    return -1/4.5 * (x - 20) ** 2 + 22
def f_3(x):
    #return -1/30 * x ** 2 + 50
    return -1/25 * (x - 18) ** 2 + 20
def f_4(x):
    return 1/100 * (x - 47) ** 2 + 1

for x in range(19):
    points.append(f_1(x))
for x in range(18, 25):
    points.append(f_2(x))
for x in range(24, 42):
    points.append(f_1(x - 5) - 2)
for x in range(41, 52):
    points.append(f_4(x))

for x in range(52, 71):
    t = x - 52
    points.append(f_1(t))
for x in range(70, 77):
    t = x - 52
    points.append(f_2(t))
for x in range(76, 105):
    t = x - 52
    points.append(f_1(t - 5) - 2)

x = []
for i in range(110):
    x.append(i)

fig, ax = fig, ax = plt.subplots(figsize=(10, 5), dpi=300)
ax.plot(x, points)
ax.grid()
plt.show()
```
